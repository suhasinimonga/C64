```python
import random

sums = []

for x in range(0,1000):
  d1 = random.randint(1,6)
  d2 = random.randint(1,6)
  sum = d1+d2
  sums.append(sum)
```

```python
import statistics as st

mean = st.mean(sums)
st_d = st.stdev(sums)
median = st.median(sums)
mode = st.mode(sums)

print("Mean:",mean)
print("Median",median)
print("Mode:",mode)
print("St_d:",st_d)
```

```python
st_d1 = []
st_d2 = []
st_d3 = []

for x in sums:
  dev = abs(mean - x)

  if(dev < st_d):
    st_d1.append(x)
  if(dev < 2*st_d):
    st_d2.append(x)
  if(dev < 3*st_d):
    st_d3.append(x)
```

```python
st_p1 = (len(st_d1) * 100.0)/ len(sums)
st_p2 = (len(st_d2) * 100.0)/ len(sums)
st_p3 = (len(st_d3) * 100.0)/ len(sums)

print("First step %:",st_p1)
print("Second step %:",st_p2)
print("Third step %:",st_p3)
```

```python
import plotly.figure_factory as pff
import plotly.graph_objects as pgo

figure = pff.create_distplot([sums],["Result"])

figure.add_trace(pgo.Scatter(x=[mean, mean], y=[0, 0.17], mode="lines", name="MEAN"))
figure.add_trace(pgo.Scatter(x=[mean - st_d, mean - st_d], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 1"))
figure.add_trace(pgo.Scatter(x=[mean + st_d, mean + st_d], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 1"))
figure.add_trace(pgo.Scatter(x=[mean - 2*st_d, mean - 2*st_d], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 2"))
figure.add_trace(pgo.Scatter(x=[mean + 2*st_d, mean + 2*st_d], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 2"))
figure.add_trace(pgo.Scatter(x=[mean - 3*st_d, mean - 3*st_d], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 3"))
figure.add_trace(pgo.Scatter(x=[mean + 3*st_d, mean + 3*st_d], y=[0, 0.17], mode="lines", name="STANDARD DEVIATION 3"))

figure.show()
```
